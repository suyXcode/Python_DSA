# Sorting Algorithm

## 🔢 What is Sorting in Computer Science?

- Sorting is the process of arranging data in a specific order — usually ascending (small → large) or descending (large → small).

- In programming and data structures, sorting organizes elements of a list, array, or dataset so they can be searched, processed, or analyzed efficiently. 🚀

📌 Example

- Suppose we have a list of numbers:

`[8, 3, 1, 6, 2]`

- Ascending Order Sorting:

`[1, 2, 3, 6, 8]`

- Descending Order Sorting:

`[8, 6, 3, 2, 1]`



## 💻 Example in Python

````python
numbers = [8, 3, 1, 6, 2]

# Ascending order
numbers.sort()
print(numbers)

# Descending order
numbers.sort(reverse=True)
print(numbers)
````

## ⚡ Why Sorting is Important

- Sorting is used in many computer science problems:

    - 🔎 Searching algorithms (Binary Search requires sorted data)
    - 📊 Data analysis
    - 🗂 Database queries
    - 🤖 Machine Learning preprocessing

## 💻 Efficient data management

- Since you work with data science and ML projects, sorting often helps in:
    - ranking predictions
    - preprocessing datasets
    - organizing results.


## 📚 Common Sorting Algorithms

| Algorithm      | Time Complexity | Idea                                 |
| -------------- | --------------- | ------------------------------------ |
| Bubble Sort    | O(n²)           | Repeatedly swap adjacent elements    |
| Selection Sort | O(n²)           | Select smallest element and place it |
| Insertion Sort | O(n²)           | Insert elements in correct position  |
| Merge Sort     | O(n log n)      | Divide and merge                     |
| Quick Sort     | O(n log n)      | Pivot partition method               |
| Heap Sort      | O(n log n)      | Use heap data structure              |

````python
nums = [4, 8, 2, 10, 3]

max_val = float('-inf')

for num in nums:
    if num > max_val:
        max_val = num

print(max_val)
````

````python
nums = [4, 8, 2, 10, 3]

min_val = float('inf')

for num in nums:
    if num < min_val:
        min_val = num

print(min_val)
````

# Count Sort 

````python
def countSort(a):
    min= a[0]
    max=a[0]
    for n in a:
        if n<min:
            min=n
        if n>max:
            max=n
    freq=[0]*(max-min+1)
    for n in a:
        freq[n-min]+=1
    j=0 
    for i in range(len(freq)):
        while freq[i]>0:
            a[j]=i+min
            j+=1
            freq[i]-=1
    return a
````

````python
a=[8,10,9, 3, 1,2, 6, 2,8]
print(countSort(a))
````
