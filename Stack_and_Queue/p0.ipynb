# Stack And Queue

```python
# simple stack formation program to add element and delete element

def stack():
    stack = []
    while True:
        print("1. Push an element")
        print("2. Pop an element")
        print("3. Display the stack")
        print("4. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            element = input("Enter the element to push: ")
            stack.append(element)
            print(f"{element} has been pushed to the stack.")
        
        elif choice == 2:
            if not stack:
                print("Stack is empty. Cannot pop an element.")
            else:
                popped_element = stack.pop()
                print(f"{popped_element} has been popped from the stack.")
        
        elif choice == 3:
            if not stack:
                print("Stack is empty.")
            else:
                print("Stack elements:")
                for elem in reversed(stack):
                    print(elem)
        
        elif choice == 4:
            print("Exiting the program.")
            break
        
        else:
            print("Invalid choice. Please try again.")
```

```python
# Run the stack program
if __name__ == "__main__":
    stack()
    
```

# Queue

```python
# Queue formation program to add element and delete element
def queue():
    queue = []
    while True:
        print("1. Enqueue an element")
        print("2. Dequeue an element")
        print("3. Display the queue")
        print("4. Exit")
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            element = input("Enter the element to enqueue: ")
            queue.append(element)
            print(f"{element} has been enqueued to the queue.")
        
        elif choice == 2:
            if not queue:
                print("Queue is empty. Cannot dequeue an element.")
            else:
                dequeued_element = queue.pop(0)
                print(f"{dequeued_element} has been dequeued from the queue.")
        
        elif choice == 3:
            if not queue:
                print("Queue is empty.")
            else:
                print("Queue elements:")
                for elem in queue:
                    print(elem)
        
        elif choice == 4:
            print("Exiting the program.")
            break
        
        else:
            print("Invalid choice. Please try again.")
```

```python
# Run the queue program
if __name__ == "__main__":
    queue()
```
