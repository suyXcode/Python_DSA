# Linked List in Python 🐍🔗

- A Linked List is a linear data structure where each element (called a node) stores:

    1. Data
    2. Reference (pointer) to the next node

- Unlike arrays/lists, linked lists are not stored in continuous memory.

## 📌 Structure of a Node

-Each node looks like this:

```bash
+------+------+
| data | next |
+------+------+
```
**Example:**

    - 10 -> 20 -> 30 -> None
    - 10 points to 20
    - 20 points to 30
    - 30 points to None (end of list)

````python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
````

````python
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
````

````python
# Connect nodes
n1.next = n2
n2.next = n3
````

````python
temp = n1
while temp:
    print(temp.data,end="-:>")
    temp = temp.next
print("None")
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at end
    def append(self, data):

        # Create new node
        new_node = Node(data)

        # If list empty
        if self.head is None:
            self.head = new_node
            return

        # Traverse till last node
        temp = self.head

        while temp.next:
            temp = temp.next

        # Connect last node to new node
        temp.next = new_node

    # Print linked list
    def display(self):

        temp = self.head

        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next

        print("None")


# Create linked list
ll = LinkedList()

# Insert data
ll.append(10)
ll.append(20)
ll.append(30)

# Display list
ll.display()
````

## 📌 Advantages of Linked List

- ✅ Dynamic size
- ✅ Easy insertion/deletion
- ✅ No memory wastage

## 📌 Disadvantages

- ❌ Extra memory for pointers
- ❌ Slow access compared to arrays
- ❌ Cannot directly access index like arrays

## 🔥 Types of Linked Lists
**1. Singly Linked List**
```10 -> 20 -> 30```
**2. Doubly Linked List**
```None <- 10 <-> 20 <-> 30 -> None```

- Each node has:

- previous pointer
- next pointer

**3. Circular Linked List**

- Last node points back to first node.
```
10 -> 20 -> 30
^           |
|___________|
```

## 📌 Time Complexity
```
Operation	               Complexity
Insert at beginning         	O(1)
Insert at end	                O(n)
Delete	                        O(n)
Search                      	O(n)
```
