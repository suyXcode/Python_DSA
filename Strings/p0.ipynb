**Valid Parentheses**

```csharp
def isValid(s):
    st=[]
    mp={')':'(',']':'[','}':'{'}
    for c in s:
        if c in mp:
            if not st or st[-1]!=mp[c]:
                return False
            st.pop()
        else:
            st.append(c)
    return not st
```

**Reverse Linked List**

```csharp
def reverseList(head):
    prev=None
    cur=head
    while cur:
        nxt=cur.next
        cur.next=prev
        prev=cur
        cur=nxt
    return prev
```

**Longest Substring**

```csharp
def lengthOfLongestSubstring(s):
    st=set()
    l=res=0
    for r in range(len(s)):
        while s[r] in st:
            st.remove(s[l]); l+=1
        st.add(s[r])
        res=max(res,r-l+1)
    return res
```

**Minimum Window Substring**

```csharp
from collections import Counter

def minWindow(s,t):
    need=Counter(t)
    have={}
    match=0
    l=0
    res=[-1,-1]
    res_len=float('inf')
    
    for r in range(len(s)):
        c=s[r]
        have[c]=have.get(c,0)+1
        
        if c in need and have[c]==need[c]:
            match+=1
        
        while match==len(need):
            if (r-l+1)<res_len:
                res=[l,r]; res_len=r-l+1
            
            have[s[l]]-=1
            if s[l] in need and have[s[l]]<need[s[l]]:
                match-=1
            l+=1
    
    l,r=res
    return s[l:r+1] if res_len!=float('inf') else ""
```
