# 🧠 What is an Array in Python?

- An array is a collection of elements of the same data type, stored in contiguous memory locations.
- **📌 Key idea:**
    - All elements → same type
    - Faster & memory-efficient than lists
    - More “low-level” than Python lists
- **👉 In Python, lists are not true arrays.**

## ✅ Way 1: Using array module (REAL Python Array)

**Python provides a standard library module called `array`.**

**🔹 Syntax**

````python
from array import array
````

````python

arr = array('i', [10, 20, 30, 40])

print(arr)
````

# 🧾 Type Codes in array module

| Code  | Data Type     |
| ----- | ------------- |
| `'i'` | Integer       |
| `'f'` | Float         |
| `'d'` | Double        |
| `'b'` | Signed char   |
| `'B'` | Unsigned char |

---

**Example : Float Array**

```python
from array import array

arr = array('f', [1.5, 2.5, 3.5])
print(arr)
```




````python
arr = array('f', [1.5, 2.5, 3.5])
print(arr)
````

## ⚠️ Important Rule (Very Common Interview Question)

**❌ You CANNOT mix data types**

```python
array('i', [10, 20.5, 30])  # ❌ ERROR


```

**✔️ Reason: array enforces homogeneous data**

## 🔥 Comparison: Array vs List

| Feature   | Array                | List          |
| --------- | -------------------- | ------------- |
| Data type | Same only            | Mixed allowed |
| Memory    | Less                 | More          |
| Speed     | Faster               | Slower        |
| Built-in  | Needs `array` module | Default       |
| Math use  | Limited              | Flexible      |


## 🎯 When should you use arrays?

- **✔️ When:**

    - You need numeric data
    - Memory efficiency matters
    - You want C-like behavior

## ❌ Avoid arrays when:
- You need mixed data
- You need flexibility



## 🧠 Final Summary (Exam Ready ✅)

- An array in Python is a collection of elements of the same data type stored in contiguous memory locations. Python supports arrays using the array module, which is more memory efficient than lists but less flexible.

## 🧠 What is NumPy?

- NumPy (Numerical Python) is a high-performance numerical computing library in Python.

- At its core is the ndarray (N-dimensional array) — a true array, unlike Python lists.


## 📦 Installing & Importing NumPy
```python
pip install numpy # install library

import numpy as np # importing the module
```

````python
import numpy as np
````

## 🔹 Creating Arrays in NumPy

**1️⃣ From a Python list**

```python 
import numpy as np

arr = np.array([10, 20, 30, 40])
print(arr)
print(type(arr))

```

**✔️ Output type → numpy.ndarray**


**2️⃣ Array with specific data type**

```python
arr = np.array([1, 2, 3], dtype=float)
print(arr)
```

**3️⃣ Zeros array**
```python
np.zeros(5)

np.zeros((2, 3))
```

**4️⃣ Ones array**
```python
np.ones((3, 2))
```

**5️⃣ Range-based arrays**
```python
np.arange(1, 10, 2)

np.linspace(0, 1, 5)
```

**6️⃣ Random arrays**

np.random.rand(3)

np.random.randint(1, 100, size=(2, 3))


## ⚡ NumPy vs Python List (🔥 Interview Favorite) 👍

| Feature   | NumPy Array | Python List |
| --------- | ----------- | ----------- |
| Data type | Same only   | Mixed       |
| Speed     | ⚡ Very fast | Slower      |
| Memory    | Efficient   | Heavy       |
| Math ops  | Vectorized  | Loop-based  |
| ML / DS   | ✅ Yes       | ❌ No        |


## 🎯 When to Use NumPy?

**✅ Use NumPy when:**
- Data Science / ML / AI
- Matrix & vector math
- Large numerical datasets
- Performance matters

**❌ Avoid when:**

- Text-heavy data
- Heterogeneous objects

## 🧠 One-Line Definition (Exam Ready)

- NumPy array is a homogeneous, fixed-size, multidimensional data structure that allows fast numerical computation using vectorized operations.
