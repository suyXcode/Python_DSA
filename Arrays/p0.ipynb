```python
#type 1
import array
val  = array.array('i', [1, 2, 3, 4, 5])
print(val)
```

```python
for x in val:
    print(x,end=" ")
```

```python
#Type 2
import array as arr
vals = arr.array('i', [10, 20, 30, 40, 50])
print(vals)
for y in vals:
    print(y,end=" ")
```

```python
# type 3
from array import *
valss = array('i', [100, 200, 300, 400, 500])
print(valss)
print("==================================")
for z in valss:
    print(z,end=" ")
```

```python
value = array('i', [5,10,15,20,25])
copyArray = array(value.typecode, (a for a in value))
print(copyArray)
print("\n==================================")
print("Array Copied Successfully")
print("==================================")
print("\nOriginal array: ", end="")
for i in value:
    print(i, end=" ")
print("\nCopied array: ", end="")
for i in copyArray:
    print(i, end=" ")
```
