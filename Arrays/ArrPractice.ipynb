**Biggest And Smallest Element from Array/List**

```python
def bigandsmall(arr):
    big=arr[0]
    small=arr[0]
    for i in arr:
        if i > big:
            big=i
        elif i<small:
            small=i 
    print(f"Biggest Element in Array is :{big}")
    print(f"Smallest Element in Array is :{small}")
    
# arr = [123,435,6453,12,324,23,6556,23,1]
arr=[2,2,2,2,2]
print(f"Array is : ")
bigandsmall(arr)
```

**Sum Except Itself**

```python
def sumExceptself(arr):
    sum=0
    for i in arr:
        sum+=i 
    for i in range(len(arr)):
        arr[i]=sum-arr[i]
    return arr 
print(sumExceptself([2,3,2,2,2]))
```

**Product Except Itself**

```python
def proExceptitself(nums):
    pro=1 
    count=0 
    for n in nums:
        if n!=0:
            pro*=n
        else:
            count+=1
    if count>1:
        return [0]*len(nums)
    for i in range(len(nums)):
        if count==0:
            nums[i]=pro/nums[i]
        else:
            if nums[i]!=0:
                nums[i]=0
            else:
                nums[i]=pro
    return nums
# single zero present in arr
arr=[2,3,4,5,6,0]
print(f"Array is : {arr}")
print(proExceptitself(arr))
print("="*20)
# two or more than two zeros present in arr
arr=[2,3,0,5,6,0]
print(f"Array is : {arr}")
print(proExceptitself(arr))
print("="*20)
#   arr without zero
arr=[2,3,4,5,6,23]
print(f"Array is : {arr}")
print(proExceptitself(arr))
```

```python
def runningSum(arr):
    sum=0 
    for i in range(len(arr)):
        sum=sum+arr[i]
        arr[i]=sum
    return arr
a=[2, 3, 0, 5, 6, 0]
b=[2, 3, 4, 5, 6, 0]
print(f"Array a:{a}\nArray b:{b}")
print(f"Running Sum Array:{runningSum(a)}")
print(f"Running Sum Array:{runningSum(b)}")
```

**Reverse Array**

```python
def reverseArray(arr):
    start=0 
    end=len(arr)-1
    while start<end:
        arr[start],arr[end]=arr[end],arr[start]
        start+=1
        end-=1
    return arr
arr=[2,3,4,5,6,0]
print(f"Array is : {arr}")
print(f"Reverse Array is : {reverseArray(arr)}")
```
