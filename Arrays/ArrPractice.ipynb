```python
def bigandsmall(arr):
    big=arr[0]
    small=arr[0]
    for i in arr:
        if i > big:
            big=i
        elif i<small:
            small=i 
    print(f"Biggest Element in Array is :{big}")
    print(f"Smallest Element in Array is :{small}")
    
# arr = [123,435,6453,12,324,23,6556,23,1]
arr=[2,2,2,2,2]
print(f"Array is : ")
bigandsmall(arr)
```
