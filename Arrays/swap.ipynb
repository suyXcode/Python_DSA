# ✅ Method 1: Using Temporary Variable (Most basic)

* This is the safest and most understandable method.

**Example in Python**

````python
arr1 = [1, 2, 3]
arr2 = [4, 5, 6]

# swap
temp = arr1
arr1 = arr2
arr2 = temp

print("arr1:", arr1)
print("arr2:", arr2)
````

# ✅ Method 2: Using Python Tuple Swapping (Best and shortest ⭐)

* Python provides a direct way.
* This is the recommended method in Python 🚀


````python
arr1 = [1, 2, 3]
arr2 = [4, 5, 6]

arr1, arr2 = arr2, arr1

print(arr1)
print(arr2)
````

# ✅ Method 3: Swap elements one-by-one (when sizes same)

- Used when you want to swap individual elements.

````python
arr1 = [1, 2, 3]
arr2 = [4, 5, 6]

for i in range(len(arr1)):
    arr1[i], arr2[i] = arr2[i], arr1[i]

print(arr1)
print(arr2)
````

# 🔥 Interview Tip – Difference

| Method                    | Swaps                  |
| ------------------------- | ---------------------- |
| `arr1, arr2 = arr2, arr1` | entire array reference |
| loop method               | individual elements    |
| temp variable             | entire array reference |


# ✅ Best Method (Recommended)
* Fast, clean, Pythonic ⚡
```python
arr1, arr2 = arr2, arr1
```
