# floating point count

```python
def float_count(num):
    

    count = 0

    while num != int(num):
        num *= 10
        count += 1

    print("Count of digits after decimal:", count)

num = float(input("Enter a float number: "))
float_count(num)
```

```python
def is_prime(num):
    for i in range(2,num):
        if num%i==0:
            return False
    else:
        return True
n= int(input("Enter the number :"))
print("Number is prime  :",is_prime(n))
```

```python
# prime number series

def primeSeries(n):
    count = 0
    for i in range(1,n+1):
        if is_prime(i):
            print(i,end=" ")
            count += 1
    print("\nTotal prime numbers between 1 and", n, "are:", count)
n= int(input("Enter the number :"))
primeSeries(n)
```

```python
# nth prime number
def nprime(n):
    count = 0
    num = 1
    while count < n:
        if is_prime(num):
            count += 1
        num  += 1
    print(f"{n} Prime number : {num-1}")
    
n = int(input("Enter the Number :"))  
nprime(n)
```
