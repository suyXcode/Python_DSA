# floating point count

```python
def float_count(num):
    

    count = 0

    while num != int(num):
        num *= 10
        count += 1

    print("Count of digits after decimal:", count)

num = float(input("Enter a float number: "))
float_count(num)
```

```python
def count_float_val(num):
    num = float(input("Enter a float number: "))

    formatted = format(num, 'f')  # remove scientific notation

    if '.' in formatted:
        decimal_part = formatted.rstrip('0').split('.')[1]
        print("Count of digits after decimal:", len(decimal_part))
    else:
        print("No decimal part")
num=float(input("Enter the input:"))
```
