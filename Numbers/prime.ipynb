```python
# prime number using function (i*i<=n ) = (i<=sqrt(n) )
def isprime(num):
    if num<=1:
        return False
    i=2
    while i*i<=num:
        if num%i==0:
            return False
        i+=1
    return True

print(isprime(11))  # True
print(isprime(4))   # False
```

```python
# best approach to check prime number 
def isprime_best(num):
    if num<=1:
        return False
    if num==2 or num==3:
        return True
    if num%2==0:
        return False
    for i in range(3, int(num**0.5)+1,2):
        if num%i==0:
            return False
    return True
print(isprime_best(29))  # True
print(isprime_best(100)) # False
```

## prime number which is given range.

```python
def primeRange(n1,n2):
    c = 0
    for i in range(n1,n2+1):
        if isprime(i):
            print(i,end=" ")
            c+=1
    print("\nTotal prime numbers in the range:",c)
primeRange(1,100)
```

```python
def kthPrime(n):
    c = 0
    i = 2
    while True:
        if isprime(i):
            c+=1
        if c==n:
            return i
        i += 1
        
print(kthPrime(10))  # 29
```

```python
def alternatePrime(n1,n2):
    c = 0
    for i in range(n1,n2+1):
        if isprime(i):
            c+=1
            if c%2==0:
                print(i,end=" ")
alternatePrime(10,100)
```

```python
def alternatePrime(n1,n2):
    c = 0
    for i in range(n1,n2+1):
        if isprime(i):
            c+=1
            if c%2==1:
                print(i,end=" ")
alternatePrime(10,100) 
```

```python
from itertools import count

def kthPrime(n):
    c = 0
    for i in count(2):        # ♾️ infinite for-loop
        if isprime(i):
            c += 1
        if c == n:
            return i
print(kthPrime(10))  # 29
```
