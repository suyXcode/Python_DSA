```python
#Write a recursive program to print numbers from 1 to N. 
def printn(n):
    if n == 0:
        return
    printn(n - 1)
    print(n)

printn(5)
```

```python
#2. Write a recursive function to find the sum of first N natural numbers.  
def sum(n: int):
    if n == 0:
        return 0
    return n + sum(n - 1)

print(sum(10))
```

```python
#3. Write a recursive program to find the factorial of a given number. 
def refact(n):
    if n==0:
        return 1
    if n==1:
        return 1
    return n*refact(n-1)
print(refact(5))
print(refact(0))
print(refact(1))
```

```python
#4. Write a recursive function to calculate the power of a number (x^y).
def power(b,n):
    if n==0:
        return 1
    return b*power(b,n-1)
print(power(5,3))
```

**4️⃣ Arithmetic Progression (AP)**

Example: 2 + 5 + 8 + ... (a=2, d=3)


```python
def ap(n, a=2, d=3):
    if n == 1:
        return a
    return (a + (n - 1) * d) + ap(n - 1, a, d)
print(ap(10))
```

**5️⃣ Geometric Progression (GP)**

Example: 2 + 4 + 8 + 16 + ... (a=2, r=2)



```python
def gp(n, a=2, r=2):
    if n == 1:
        return a
    return (a * (r ** (n - 1))) + gp(n - 1, a, r)
print(gp(5))
```

```python
# Write a recursive function to reverse a given number.
def rever(n,rev=0):
    if n==0:
        return rev
    return rever(n//10, rev*10 + n%10)
print(rever(123))
        
    
```

```python
#7. Write a recursive function to count the number of digits in a given number.

def countdigit(n):
    if n<10:
        return 1
    if n<1:
        return 0
    
    return 1+countdigit(n//10)
print(countdigit(1234))
print(countdigit(0))
```

```python
# 8. Write a recursive function to find the sum of digits of a number.  
def sumofdigit(n,sum=0):
    if n==0:
        return sum
    sum += n%10
    return sumofdigit(n//10,sum)
print(sumofdigit(1234))
```

```python
#9.  Write a recursive function to find the LCM (Least Common Multiple) of two numbers. 

def lcm(a,b,maxi):
    if maxi%a==0 and maxi%b==0:
        return maxi
    return lcm(a,b,maxi+1)
print(lcm(4,6, max(4,6)))
```

```python
#11. Write a recursive function to find the sum of elements in an array.
def sumofarray(arr, n):
    if n == 0:
        return 0
    return arr[n - 1] + sumofarray(arr, n - 1)
```

```python
# 12. Write a recursivee function to find the maximum element in a list
def maxoflist(arr, n):
    if n == 1:
        return arr[0]
    return max(arr[n - 1], maxoflist(arr, n - 1))
```

```python
# 13. Write a recursive function to check whether a string is palindrome or not. 
def is_palindrome(s, start, end):
    if start >= end:
        return True
    if s[start] != s[end]:
        return False
    return is_palindrome(s, start + 1, end - 1)
```

```python
# 14. Write a recursive function to convert a decimal number to binary. 
def decimal_to_binary(n):
    if n == 0:
        return "0"
    if n == 1:
        return "1"
    return decimal_to_binary(n // 2) + str(n % 2)
```

```python
#15. Write a recursive function to find the GCD (Greatest Common Divisor) of two numbers.  
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)
```
