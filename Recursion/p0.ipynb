```python
#Write a recursive program to print numbers from 1 to N. 
def printn(n):
    if n == 0:
        return
    printn(n - 1)
    print(n)

printn(5)
```

```python
##Write a recursive program to print numbers from N to 1
def printn(n:int):
    if n == 0:
        return
    print(n)
    printn(n - 1)
    

printn(5)
```

```python
#2. Write a recursive function to find the sum of first N natural numbers.  
def sum(n: int):
    if n == 0:
        return 0
    return n + sum(n - 1)

print(sum(10))
```

```python
#3. Write a recursive program to find the factorial of a given number. 
def refact(n):
    if n==0:
        return 1
    if n==1:
        return 1
    return n*refact(n-1)
print(refact(5))
print(refact(0))
print(refact(1))
```

```python
#4. Write a recursive function to calculate the power of a number (x^y).
def power(b,n):
    if n==0:
        return 1
    return b*power(b,n-1)
print(power(5,3))
```

```python
#5. Write a recursive program to find the Nth term of the Fibonacci series.

#The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, starting typically with 0 and 1, creating the pattern: 0, 1, 1, 2, 3, 5, 8, 13

def fibb(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fibb(n-1)+fibb(n-2)
index=5
print(fibb(index))

# 0, 1, 1, 2, 3, 5, ...
# index: 0  1  2  3  4  5

    
```

**4️⃣ Arithmetic Progression (AP)**

Example: 2 + 5 + 8 + ... (a=2, d=3)


```python
def ap(n, a=2, d=3):
    if n == 1:
        return a
    return (a + (n - 1) * d) + ap(n - 1, a, d)
print(ap(10))
```

**5️⃣ Geometric Progression (GP)**

Example: 2 + 4 + 8 + 16 + ... (a=2, r=2)



```python
def gp(n, a=2, r=2):
    if n == 1:
        return a
    return (a * (r ** (n - 1))) + gp(n - 1, a, r)
print(gp(5))
```

```python
# Write a recursive function to reverse a given number.
def rever(n,rev=0):
    if n==0:
        return rev
    return rever(n//10, rev*10 + n%10)
print(rever(123))
        
    
```

```python
#7. Write a recursive function to count the number of digits in a given number.

def countdigit(n):
    if n<10:
        return 1
    if n<1:
        return 0
    
    return 1+countdigit(n//10)
print(countdigit(1234))
print(countdigit(0))
```
